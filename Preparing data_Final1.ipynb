```python
import numpy as np
import pandas as pd
import json
import datetime as dt
```

```python
data = []
#with gzip.open('C:/Users/fef/OneDrive - Ultranav/DTU/Semester 4/02806 - Social data analysis & visualization/Project/data/endomondoHR.json.gz') as f:
with open('C:/Users/faraz/OneDrive - Ultranav/DTU/Semester 4/02806 - Social data analysis & visualization/Project/data/endomondoHR_proper.json') as f:
    for l in f:
        data.append(eval(l))
```

```python
# append all data list to a single list
appended_data = []
for i in range(167783):
    appended_data.append(pd.DataFrame(data[i]))
# make a dataset based on multiple lists
appended_data = pd.concat(appended_data)
```

```python
# choose the column needed for data analysis
appended_data1 = appended_data[['longitude','latitude','sport','heart_rate','gender','timestamp','userId']].reset_index(drop=True)
```

```python
# make datetime column based on timestamp
appended_data1['datetime'] = pd.to_datetime(appended_data1['timestamp'],unit='s')
# Convert datetime column to pandas datetime
appended_data1['datetime'] = pd.to_datetime(appended_data1['datetime'])
# make a date column based on datetime
appended_data1['date'] = appended_data1['datetime'].dt.date
# convert date colun to string
appended_data1['date'] = appended_data1['date'].astype(str)
```

```python
# convert userId column to string
appended_data1['userId'] = appended_data1['userId'].astype(str)
```

```python
# sort dataset based on datetime column
appended_data2= appended_data1.sort_values(by='datetime')
```
